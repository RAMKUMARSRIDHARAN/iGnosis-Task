# Merging the both Dataset and save Dataframe as " Market_Segment"

```python
# Merge both datasets on LYLTY_CARD_NBR
Market_Segment = transaction.merge(purchase, on="LYLTY_CARD_NBR", how="left")
```

# Identifying and Including only the Unique columns 

```python
# List of columns to exclude
Include_columns = ["PREMIUM_CUSTOMER","LIFESTAGE","PROD_QTY","PROD_NAME"]  

# Iterate through each column except the excluded ones
for column in Market_Segment.columns:
    if column in Include_columns:
        # Get unique values for the column
        Unique_columns = Market_Segment[column].unique()

        print(f"Unique values for column '{column}':")
        print("-" * 100)
        print(Unique_columns.tolist())  # Convert NumPy array to list for better readability
        print("\n")  # Print a new line for better readability
```
